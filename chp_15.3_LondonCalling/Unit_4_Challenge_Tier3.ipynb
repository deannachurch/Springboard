**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Just make a plot for each borough
for b in borough_to_keep:
    df= prop_melt_borough[prop_melt_borough['Borough'] == b].copy()
    df['Year_Month'] = df['Date'].dt.strftime('%Y_%m')
    df['Average_Price'] = df['Average_Price'] / 1000000 #convert to millions
    # Plot the average price against the date
    df.plot(x='Year_Month', y='Average_Price', kind='line')

    # Remove the top and right spines
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Set the plot title and labels
    plt.title(f"Average Price in {b} Over Time")
    plt.xlabel('Date')
    plt.ylabel('Average Price (Millions of Pounds)')

    #save plot
    plt_name='img/' + b + '_plot.png'
    plt.savefig(plt_name, dpi=300)
    # Display the plot
    plt.show()
```

```python
# Calculate the start date of the two-decade period
start_date = datetime.datetime.now() - datetime.timedelta(days=365*20)

# Filter the dataframe
prop_melt_borough_2decade = prop_melt_borough[prop_melt_borough['Date'] >= start_date].copy()
prop_melt_borough_2decade.shape
```

```python
display(prop_melt_borough_2decade.head())
```

```python
#calculate average price increase per borough
# Group the data by Borough and Year_Month
avg_price=prop_melt_borough_2decade[['Borough', 'Date', 'Average_Price']].copy()
avg_price.sort_values(['Borough', 'Date'], inplace=True)
display(avg_price.head())
```

```python
# Pivot the data to have Borough as index, Year_Month as columns, and Average_Price as values
avg_price_pivot = avg_price.pivot(index='Borough', columns='Date', values='Average_Price')
display(avg_price_pivot.head())
```

```python
# Calculate the average price increase for each borough over the past two decades
avg_price_increase = avg_price_pivot.iloc[:, -1] - avg_price_pivot.iloc[:, 0]
display(avg_price_increase.head())
```

```python
# Sort the result in descending order
avg_price_increase.sort_values(ascending=False, inplace=True)

# Display the result
display(avg_price_increase)
```

```python
# Create a horizontal bar chart of avg_price_increase
avg_price_increase.sort_values().plot(kind='barh', figsize=(7, 8))

# Remove the top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the title and labels
plt.title('Average Price Increase in London Boroughs Over the Past Two Decades')
plt.xlabel('Average Price Increase (Pounds)')
plt.ylabel('Borough')

# Display the plot
plt.tight_layout()
plt.savefig('img/AvgPriceIncrease.png', dpi=300)
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(df):
    """calculate price ratio for a house for 1998 and 2018"""
    avg_price = df.groupby(df['Date'].dt.year)['Average_Price'].first()
    return avg_price.loc[1998] / avg_price.loc[2018]
```

```python
price_ratio = prop_melt_borough.groupby('Borough').apply(create_price_ratio)
price_ratio.sort_values(ascending=False, inplace=True)
print(price_ratio)
```

```python
# Create a horizontal bar chart of avg_price_increase
price_ratio.plot(kind='barh', figsize=(7, 8))

# Remove the top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set the title and labels
plt.title('Price ratio between 1998 and 2018, per Borough')
plt.xlabel('Price Ratio')
plt.ylabel('Borough')

# Display the plot
plt.tight_layout()
plt.savefig('img/PriceRatio1998_2018.png', dpi=300)
plt.show()
```
