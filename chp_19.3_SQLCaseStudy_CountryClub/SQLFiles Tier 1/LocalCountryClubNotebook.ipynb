```python
#From LocalSqlConnecion.py
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn, query):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    

    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)

    
    with conn: 
        query1 = """
        SELECT *
        FROM FACILITIES
        """
        print("2. Query all tasks")
        select_all_tasks(conn, query1)
 
 
if __name__ == '__main__':
    main()
```

```python
""" Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! """ 

with conn: 
    query1 = """ 
    SELECT 
        f.name,
        SUM(CASE 
            WHEN b.memid = 0 THEN b.slots * f.guestcost
            ELSE b.slots * f.membercost
        END) AS total_revenue
    FROM Bookings b
    JOIN Facilities f ON b.facid = f.facid
    GROUP BY f.name
    HAVING total_revenue < 1000
    ORDER BY total_revenue ASC; 
    """
    select_all_tasks(conn, query1)
```

```python
"""Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order""" 

with conn:
    query1 = """
        SELECT 
            m1.surname AS member_surname,
            m1.firstname AS member_firstname,
            m2.surname AS recommender_surname,
            m2.firstname AS recommender_firstname
        FROM members m1
        LEFT JOIN members m2 ON m1.recommendedby = m2.memid
        ORDER BY m1.surname, m1.firstname;
        """ 
    select_all_tasks(conn, query1)
```

```python
""" Find the facilities with their usage by member, but not guests  """
#This is worded very poorly...

with conn:
    query1 = """
        SELECT 
            f.name AS facility_name,
            m.surname AS member_surname,
            m.firstname AS member_firstname,
            COUNT(b.bookid) AS usage_count
        FROM Bookings b
        JOIN Members m ON b.memid = m.memid
        JOIN Facilities f ON b.facid = f.facid
        WHERE b.memid != 0
        GROUP BY f.name, m.surname, m.firstname
        ORDER BY f.name, m.surname, m.firstname;
        """ 
    select_all_tasks(conn, query1)

```

```python
""" /* Q13: Find the facilities usage by month, but not guests */ """

with conn:
    query1 = """
        SELECT 
            f.name AS facility_name,
            strftime('%Y', b.starttime) AS year,
            strftime('%m', b.starttime) AS month,
            COUNT(b.bookid) AS usage_count
        FROM Bookings b
        JOIN Facilities f ON b.facid = f.facid
        WHERE b.memid != 0
        GROUP BY f.name, year, month
        ORDER BY f.name, year, month;
        """ 
    select_all_tasks(conn, query1)
```
